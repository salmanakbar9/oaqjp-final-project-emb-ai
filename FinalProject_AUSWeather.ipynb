```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns
```

```python
url="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/_0eYOqji3unP1tDNKWZMjg/weatherAUS-2.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.count()
```

```python
df = df.dropna()
df.info()
```

```python
df.columns
```

```python
df = df.rename(columns={'RainToday': 'RainYesterday',
                        'RainTomorrow': 'RainToday'
                        })
```

```python
df = df[df.Location.isin(['Melbourne','MelbourneAirport','Watsonia',])]
df. info()
```

```python
def date_to_season(date):
    month = date.month
    if (month == 12) or (month == 1) or (month == 2):
        return 'Summer'
    elif (month == 3) or (month == 4) or (month == 5):
        return 'Autumn'
    elif (month == 6) or (month == 7) or (month == 8):
        return 'Winter'
    elif (month == 9) or (month == 10) or (month == 11):
        return 'Spring'
```

```python

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Apply the function to the 'Date' column
df['Season'] = df['Date'].apply(date_to_season)

# Drop the 'Date' column
df = df.drop(columns=['Date'])
```

```python

# Drop the target column 'RainToday' to create the feature matrix (X)
X = df.drop(columns='RainToday', axis=1)

# Assign the target column 'RainToday' to y
y = df['RainToday']
```

```python

from sklearn.model_selection import train_test_split

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python

# Detect numerical features
numeric_features = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Detect categorical features
categorical_features = X_train.select_dtypes(include=['object', 'category']).columns.tolist()
```

```python
# Scale the numeric features
numeric_transformer = Pipeline(steps=[('scaler', StandardScaler())])

# One-hot encode the categoricals
categorical_transformer = Pipeline(steps=[('onehot', OneHotEncoder(handle_unknown='ignore'))])
```

```python

from sklearn.compose import ColumnTransformer

# Combine the transformers into a single preprocessing step
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),  # Apply to numeric features
        ('cat', categorical_transformer, categorical_features)  # Apply to categorical features
    ]
)
```

```python

from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier

# Create the pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),  # Apply preprocessing
    ('classifier', RandomForestClassifier(random_state=42))  # Apply Random Forest classifier
])
```

```python
param_grid = {
    'classifier__n_estimators': [50, 100],
    'classifier__max_depth': [None, 10, 20],
    'classifier__min_samples_split': [2, 5]
}
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True)
```

```python

from sklearn.model_selection import GridSearchCV

# Define the parameter grid for RandomForestClassifier
param_grid = {
    'classifier__n_estimators': [100, 200],  # Number of trees in the forest
    'classifier__max_depth': [None, 10, 20],  # Maximum depth of the trees
    'classifier__min_samples_split': [2, 5],  # Minimum samples required to split a node
    'classifier__min_samples_leaf': [1, 2]   # Minimum samples required at a leaf node
}

# Instantiate GridSearchCV
grid_search = GridSearchCV(
    pipeline,  # The pipeline to optimize
    param_grid,  # The parameter grid to search
    cv=5,  # 5-fold cross-validation
    scoring='accuracy',  # Evaluation metric
    verbose=2  # Verbosity level for logging
)

# Fit GridSearchCV to the training data
grid_search.fit(X_train, y_train)
```

```python
print("\nBest parameters found: ", grid_search.best_params_)
print("Best cross-validation score: {:.2f}".format(grid_search.best_score_))
```

```python

# Calculate the model's score on the test set
test_score = grid_search.score(X_test, y_test)

# Print the test set score with 2 decimal places
print("Test set score: {:.2f}".format(test_score))
```

```python

y_pred = grid_search.predict(X_test)
```

```python

from sklearn.metrics import classification_report

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

conf_matrix = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix)
disp.plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```
